```python
dataset = pd.read_csv(r'C:\Users\Admin\Downloads\6th, 7th- ANN THEORY, Installation\6th, 7th- ANN THEORY, Installation\CHURN MODELING- TF\Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
print(X)
print(y)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
print(X)
```

```python
# One Hot Encoding the "Geography" column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X)) 
print(X)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Part 2 - Building the ANN

# Initializing the ANN
ann = tf.keras.models.Sequential()#establish layer one after another
```

```python
ann.add(tf.keras.layers.Dense(units=6))#input layer
```

```python
# Adding the input layer and the first hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))#layer 1

ann.add(tf.keras.layers.Dense(units=5, activation='relu'))#layer 2

ann.add(tf.keras.layers.Dense(units=4, activation='relu'))#layer 3
```

```python
# Adding the output layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))#output layer
```

```python
# Part 3 - Training the ANN

# Compiling the ANN
ann.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Training the ANN on the Training set
ann.fit(X_train, y_train, batch_size = 32, epochs = 5)
```

```python
# Part 4 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test, y_pred)
print(ac)
```
